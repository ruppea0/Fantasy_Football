```python
def create_dataframes(player_data):
    '''
    The player_data is a list of dictionaries. Each dictionary has a key 'tm' and a key 'stats'.
    The value of 'tm' is a dictionary with a key 'dspNm' and a value which is the team name.
    The value of 'stats' is a list of dictionaries. Each dictionary has a key 'type' and a key 'athlts'.
    The value of 'type' is a string which is the type of the stats for a player.
    The value of 'athlts' is a list of dictionaries. Each dictionary has a key 'athlt' and a key 'stats'.
    The value of 'athlt' is a dictionary with a key 'dspNm' and a value which is the player name.
    The value of 'stats' is a list of dictionaries. Each dictionary has a key 'name' and a key 'value'.
    The value of 'name' is a string which is the name of the stat.
    The value of 'value' is a string which is the value of the stat.
    '''
    restricted_types = ['defensive', 'punting', 'puntReturns', 'kickReturns', 'interceptions']
    dataframe_dict = {}
    # for each team
    for item in player_data:
        team = item['tm']['dspNm']
        stats = item['stats']
        # for each type of player
        for player_stat in stats:
            players = player_stat['athlts']
            if player_stat['type'] not in restricted_types:
                if player_stat['type'] not in dataframe_dict:
                    dataframe_dict[player_stat['type']] = pd.DataFrame(columns=['team', 'player'] + player_stat['keys'])
                # for each player in that type
                for player in players:
                    player_name = player['athlt']['dspNm']
                    numeric_data = player['stats']
                    to_add = []
                    to_add.append(team)
                    to_add.append(player_name)
                    # for their stats
                    for number in numeric_data:
                        to_add.append(number)
                    dataframe_dict[player_stat['type']] = pd.concat([dataframe_dict[player_stat['type']], pd.DataFrame([to_add], columns=['team', 'player'] + player_stat['keys'])])

    return dataframe_dict
```

```python
for i in range(1, 19):
    print('Week ' + str(i))
    user_agent = {'User-agent': 'Mozilla/5.0',
              'referer': 'https://www.espn.com/'}
    soup = BeautifulSoup(requests.get('https://www.espn.com/nfl/scoreboard/_/week/' + str(i) + '/year/2021/seasontype/2',
                                  headers=user_agent).text, 'html.parser')
    
    links = []
    for item in soup.findAll('a'):
        # if the text of the link is 'Box Score', then we have a link to a box score
        if item.text == 'Box Score':
            # print(item)
            # get the link
            link = item.get('href')
            # add the link to the list of links
            link = 'https://www.espn.com' + link
            links.append(link)
    
    dataframe_dict = {}
    for link in links:
        response = requests.get(link, headers=headers, cookies=cookies)
        box_score = BeautifulSoup(response.text, 'html.parser')
        script_I_want = None
        for line in box_score.findAll('script'):
            for item in line:
                if ('window['  in str(item)):
                    script_I_want = line
        # split on window
        scripts = str(script_I_want).split('window')
        # cut scripts[-1] off at the first { and the last }
        json_data = scripts[-1].split('{', 1)[1]
        json_data = json_data.rsplit('}', 1)[0]
        json_data = '{' + json_data + '}'
        data = json.loads(json_data)
        player_data = data['page']['content']['gamepackage']['bxscr']
        if len(dataframe_dict) == 0:
            dataframe_dict = create_dataframes(player_data)
        else:
            new_data = create_dataframes(player_data)
            for key in dataframe_dict:
                dataframe_dict[key] = pd.concat([dataframe_dict[key], new_data[key]])

    os.mkdir('data/2021/wk' + str(i))
    for key in dataframe_dict:
        dataframe_dict[key].to_csv('data/2021/wk' + str(i) + '/'+ key + '.csv', index=False)
    
    time.sleep(15)
```

```python
def find_key_path(dictionary, target_key, current_path=None):
    if current_path is None:
        current_path = []

    for key, value in dictionary.items():
        if key == target_key:
            return current_path + [key]

        if isinstance(value, dict):
            found = find_key_path(value, target_key, current_path + [key])
            if found:
                return found

    return None
```
